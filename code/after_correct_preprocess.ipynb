# #3 Load Stopword Vi-En Join và xoá trước khi segmentation

## Xoá dấu câu

```python
import re
df = df[df['corrected_text'].notnull()]
pattern = '[^a-z0-9A-Z_ÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚĂĐĨŨƠàáâãèéêìíòóôõùúăđĩũơƯĂẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼỀỀỂưăạảấầẩẫậắằẳẵặẹẻẽềềểỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪễếệỉịọỏốồổỗộớờởỡợụủứừỬỮỰỲỴÝỶỸửữựỳýỵỷỹ]+'
def remove_special_character(sentences):
    sentences = sentences.replace('  ', ' ');
    return re.sub(pattern, ' ', sentences).lower()
df['ac_remove_special_character']=df['corrected_text'].apply(remove_special_character)
# df['ac_remove_dot'] = df['ac_remove_special_character'].str.replace('[^\w\s]','')
df
```

## Load stopword Vi-En

```python
import nltk #stopword library
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_en = stopwords.words('english')
with open('/content/drive/MyDrive/CK_Data/vietnamese-stopwords_remake.txt', 'r', encoding='utf-8') as file:
    stop_vi = file.read().splitlines()
stop_words = stop_en + stop_vi
```

```python
stop_words
```

```python
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
df['ac_remove_stopword'] = df['ac_remove_special_character'].apply(remove_stopwords)
df

```

```python
df['ac_remove_stopword'][100:200]
```

```python
def word_segmentation(sentence):
  word_segmented_text = rdrsegmenter.tokenize(sentence)
  word_segmented_text = ' '.join([' '.join(x) for x in word_segmented_text])
  print(word_segmented_text)
  return word_segmented_text
df['word_segmentation'] = df['ac_remove_stopword'].apply(word_segmentation)
df
```

```python
from underthesea import word_tokenize
# loại bỏ cột bằng rỗng hoặc
df = df[df['word_segmentation'] != '']
df = df.dropna(subset=['word_segmentation'])
df.drop_duplicates(subset="word_segmentation", inplace=True, keep=False)

df['final_preprocess'] = df['word_segmentation'].apply(lambda x: len(word_tokenize(x)))

print(df.shape)

# loại bỏ cột có số lượng token tự nhỏ hơn 1 và lớn hơn 150
df = df[df['final_preprocess'] >= 2]
df = df[df['final_preprocess'] <= 150]
df_final = df.loc[:, ['star.1', 'corrected_text', 'final_preprocess']]
print(df_final.shape)
```
