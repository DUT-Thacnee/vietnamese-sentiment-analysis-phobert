#### #step 1:Chuyển về chữ thường và loại bỏ kí tự đặc biêt

```python
pattern = '[^a-z0-9A-Z_ÀÁÂÃÈÉÊÌÍÒÓÔÕÙÚĂĐĨŨƠàáâãèéêìíòóôõùúăđĩũơƯĂẠẢẤẦẨẪẬẮẰẲẴẶẸẺẼỀỀỂưăạảấầẩẫậắằẳẵặẹẻẽềềểỄỆỈỊỌỎỐỒỔỖỘỚỜỞỠỢỤỦỨỪễếệỉịọỏốồổỗộớờởỡợụủứừỬỮỰỲỴÝỶỸửữựỳýỵỷỹ]+'
def lower_word(sentences):
    sentences = sentences.replace('  ', ' ');
    return re.sub(pattern, ' ', sentences).lower()
df['lowercase']=df['comments'].apply(lower_word)
df
```

#### #step 2: xoá các dấu chấm câu

```python
df['remove_dot'] = df['lowercase'].str.replace('[^\w\s]','')
df
```

#### #Step 3 Xoá các chữ dính số

```python
# def remove_number(input_text):
#   words = input_text.split()
#   filtered_words = []
#   for word in words:
#     if not re.match(r'\w*\d+\w*', word) and not word.isdigit():
#       filtered_words.append(word)
#   return ' '.join(filtered_words)

def remove_number(input_text):
    words = input_text.split()
    filtered_words = []
    for word in words:
        if not re.search(r'\d', word):
            filtered_words.append(word)
    return ' '.join(filtered_words)

df['remove_number'] = df['remove_dot'].apply(remove_number)
```

#### #Step 4 thay thế các từ viết tắt thông dụng

```python
with open(r'D:\CK-20250301T053445Z-001\CK\replace_acronyms.txt', 'r', encoding='utf-8') as file: #thêm chữ r để nó hiểu là chuỗi thô và không thoát ký tự '\'
    corrections = file.read().splitlines()

corrections
lookup_dict = {}
for correction in corrections:
    incorrect, correct = correction.split(':')
    if correct:
      lookup_dict[incorrect] = correct
    else:
      lookup_dict[incorrect] = incorrect

def replace_acronyms(input_text):
    words = input_text.split()
    new_words = []
    for word in words:
      if word in lookup_dict:
        word = lookup_dict[word]
      new_words.append(word)
    new_text = " ".join(new_words)
    return new_text

df['replace_acronyms'] = df['remove_number'].apply(replace_acronyms)
df
```

```python
df = df.drop_duplicates(subset=['replace_acronyms'])
df
```

#### #step 5 lọc các từ viết sai chính tả chủ yếu

```python
def corrected_text_to_array():
    array_words = []
    for sentence in df['replace_acronyms']:
        words = sentence.split()
        array_words.extend(words)

    return array_words

array_corrected_text = corrected_text_to_array()
frequency_dist = nltk.FreqDist(array_corrected_text)
single_frequency_words = frequency_dist.hapaxes()
def remove_single_frequency_words(sentence):
    words = sentence.split()

    filtered_words = []
    for word in words:
        if word not in single_frequency_words:
            filtered_words.append(word)

    filtered_sentence = " ".join(filtered_words)

    return filtered_sentence
single_frequency_words
```

```python
with open('word_frequencies.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Word', 'Frequency'])
    for word in frequency_dist:
        writer.writerow([word, frequency_dist[word]])
```

```python
df['replace_incorrect_word'] = df['replace_acronyms'].apply(remove_single_frequency_words)
df['replace_incorrect_word'].head(10)
```

```python
from transformers import pipeline # for using model in huggingface
#Load model
corrector = pipeline("text2text-generation", model="bmd1905/vietnamese-correction")
```

```python
total_record = df.shape[0]
def spell_corrector(text):
    spell_corrector.count +=1
    generated_text = corrector(text, max_length=256)
    corrected_text = generated_text[0]['generated_text']
    print(f"[{spell_corrector.count}/{total_record}] {corrected_text}")
    return pd.Series({'generated_text': generated_text, 'corrected_text': corrected_text})
```

```python
spell_corrector.count = 0
df[['generated_text', 'corrected_text']] = df['replace_incorrect_word'].apply(spell_corrector)
df.to_csv(rf"D:\CK-20250301T053445Z-001\CK\data_done_1.csv", index= False, encoding='utf-8-sig')
```
